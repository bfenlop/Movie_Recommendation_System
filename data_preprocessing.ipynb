```python
items = items.drop(columns = ['day', 'month', 'year', 'IMDB_url', 'unknown', 'OMDb_Rating', 'OMDb_imdbRating', 'OMDb_Metascore', 'OMDb_Rated', 'OMDb_BoxOffice'])
```

```python
print(items.isna().sum())
```

```python
items.dropna(inplace = True)
```

```python
items.shape
```

```python
import json
items['OMDb_Ratings'] = items['OMDb_Ratings'].str.replace("'", '"')
# Función para convertir las cadenas JSON en listas de diccionarios
def parse_ratings(ratings):
    if isinstance(ratings, str):  # Asegurar que sea una cadena
        try:
            return json.loads(ratings)  # Convertir la cadena JSON en objeto Python
        except json.JSONDecodeError:
            return []  # Devolver una lista vacía si la conversión falla
    return ratings

# Convertir la columna `OMDb_Ratings`
items['OMDb_Ratings'] = items['OMDb_Ratings'].apply(parse_ratings)

# Función para extraer valores
def extract_rating(source, ratings):
    for rating in ratings:
        if rating['Source'] == source:
            return rating['Value']
    return None  # Si no se encuentra el source

# Crear nuevas columnas a partir de los valores
items['IMDB_Score'] = items['OMDb_Ratings'].apply(lambda x: extract_rating('Internet Movie Database', x))
items['RottenTomatoes'] = items['OMDb_Ratings'].apply(lambda x: extract_rating('Rotten Tomatoes', x))
items['Metacritic'] = items['OMDb_Ratings'].apply(lambda x: extract_rating('Metacritic', x))

# Eliminar la columna original si no es necesaria
items = items.drop(columns=['OMDb_Ratings'])
```

```python
# Filtrar filas donde la longitud de 'OMDb_Year' es mayor a 4, las filas con año de longitud mayor a 4 corresponden a series (ejemplo: 2012-2016). Filtramos para que la base de datos tengan solo peliculas
filtered_items = items[items['OMDb_Year'].str.len() > 4]
items = items.drop(filtered_items.index)
```

```python
# Eliminar 'min' y convertir a número
items['OMDb_Runtime'] = items['OMDb_Runtime'].str.replace(' min', '').astype(int)
items['IMDB_Score'] = items['IMDB_Score'].str.replace('/10', '').astype(float)
items['OMDb_imdbVotes'] = items['OMDb_imdbVotes'].str.replace(',', '').astype(int)
items['OMDb_Year'] = items['OMDb_Year'].astype(int)
```

```python
print(items.isna().sum())
```

```python
items = items.drop(columns = ['RottenTomatoes', 'Metacritic', 'OMDb_Genre', 'OMDb_Released'])
```

```python
items = items.drop_duplicates(subset=['title'])
```
